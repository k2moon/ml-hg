## 결정 계수 ($ R^2$) 
---
- 회귀에서 평가 값
- 분류 :  정확도
- 타깃의 평균 정도를 예측하는 수준이면 분모 분자가 비슷해 져서 결정계수는 0에 가까워짐
- 예측이 타깃에 가까워지면 분자가 0에 가까워져서 결정계소는 1에 가까워짐
- 결정계수 0.99면 좋은 값이지만 정확도 처럼 얼마나 좋은지 이해하기 어려움

### 다른 평가 값
---
- 타깃과 예측의 절대값 오차의 평균 MAE
- 타깃과 예측의 평균 제곱오차 MSE
- MSE 제곱근 오차 : RMSE
- sklearn.metrics 패키지 사용 해서 계산
- MAE : mean_absolute_error()
- MSE : mean_squared_error()

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knr = KNeighborsRegressor()
# k-최근접 이웃 회귀 모델을 훈련합니다
knr.fit(train_input, train_target)
```

```python
knr.score(test_input, test_target)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
# 테스트 세트에 대한 예측을 만듭니다
test_prediction = knr.predict(test_input)
# 테스트 세트에 대한 평균 절댓값 오차를 계산합니다
mae = mean_absolute_error(test_target, test_prediction)
print(mae)
```

> **결과에서 예측이 평균적으로 19g 정도 타깃과 다르다는 뜻**

## 과대적합 vs 과소적합
---

### 과대적합 overfitting
---
- 보통은 훈련 세트의 점수가 조금 더 높게 나옴 
- 훈련 세트에서는 높은 점수였으나 테스트 세트에서 점수가 굉장히 낮은 모델


### 과소적합 underfitting
---
- 과대적합의 반대이거나 두 점수 모두 낮은 모델
- 훈련 세트와 테스트 세트의 크기가 매우 작은 경우에도 발생


```python
print(knr.score(train_input, train_target))
```

> **현재 과소 적합 상태 임**

### 복잡도 높이기
---
- 과소 적합을 해결 하려면 모델의 복잡도를 높이며 됨
- k-최근접이웃 알고리즘에서는 k의 값을 줄이는 것이 복작도를 높이는 것임

```python
# 이웃의 갯수를 3으로 설정합니다
knr.n_neighbors = 3
# 모델을 다시 훈련합니다
knr.fit(train_input, train_target)
print(knr.score(train_input, train_target))
```

```python
print(knr.score(test_input, test_target))
```

## 확인문제
---
- k값을 1, 5, 10으로 바꿔가며 복잡도 확인 해보기

```python
# k-최근접 이웃 회귀 객체를 만듭니다
knr = KNeighborsRegressor()
# 5에서 45까지 x 좌표를 만듭니다
x = np.arange(5, 45).reshape(-1, 1)

# n = 1, 5, 10일 때 예측 결과를 그래프로 그립니다.
for n in [1, 5, 10]:
    # 모델 훈련
    knr.n_neighbors = n
    knr.fit(train_input, train_target)
    # 지정한 범위 x에 대한 예측 구하기 
    prediction = knr.predict(x)
    # 훈련 세트와 예측 결과 그래프 그리기
    plt.scatter(train_input, train_target)
    plt.plot(x, prediction)
    plt.title('n_neighbors = {}'.format(n))    
    plt.xlabel('length')
    plt.ylabel('weight')
    plt.show()
```

### 키워드
---
- 회귀 : 임의의 수치를 예측하는 문제, 타깃도 임의 수치
- k-최근접 이웃 회귀 : 가장 가까운 이웃 샘플을 찾고 이 샘플들의 타깃값을 평균하여 예측
- 결정계수 : 회귀 문제의 대표적인 성능 측정 지표, 1에 가까울 수록 좋고, 0에 가까울 수록 성능이 나쁜 모델
- 과대 적합 : 모델이 훈련 세트에만 성능이 높은 모델, 복잡도를 낮춰서 모델을 만들어야 함

---

---

## 손실 함수 loss function
---
- 샘플 하나에 대한 손실
- 비용 함수 cost function : 모든 샘플의 손실 함수의 합
- 손실은 작을 수록 좋으나 어떤 값이 최소값인지 모름


### 경사 하강법과 손실 함수와 관계
---
- 경사 하강법 : 어떤 손실 함수가 정의되었을 때 손실함수의 값이 최소가 되는 지점을 찾아가는 방법
- 손실 함수 : 모델을 통해 예측한 값과 실제 예측 사이의 차이를 함수로 정의한 것

### 선형 회귀 분석의 손실 함수
---
- 제곱 오차 , SE
- 1개의 최적해를 갖는 이차 함수 Convex (볼록 함수) 형태
- 학습률 Learning Rate로 얼마나 이동할 지 결정

> 예측값 **$ \hat{y} = Wx + b $**

> 제곱 오차 **$ SE = \frac {1}{2} (y - \hat{y})^2 $**

> 전체식 **$ W^2x^2 + 2Wxb + b^2 -2Wxy -2yb + y^2 $**

> 가중치 편미분(경사하강법) **$ \frac{\partial SE}{\partial W} = -(y - \hat{y})\cdot x  $**

> 학습률 추가 **$  \frac{\partial SE}{\partial W} = - {\eta} \cdot (y - \hat{y})\cdot x  $**

> 가중치 업데이트 **$ W' = W - \frac{\partial SE}{\partial W} = W + {\eta} \cdot (y - \hat{y})\cdot x  $**

### 이진 로지스틱 회귀 비용함수 재정의
---
- 지수함수의 역함수인 Log 연산 적용
- 로지스틱 손실 함수 또는 이진 크로스엔트로피 손실 함수
- non-convex에서 convex 형태로 변경
- 경사 하강법 적용 가능 해짐

> **Z = H(x) = Wx + b**

> **$ S(Z) = \frac{1}{(1 + exp^-z)} $**

> **$ C(H(x), y) = \begin{cases}-log(H(x)) & : y = 1\\ -log(1-H(x))  & : y = 0\end{cases} $**

## SGDClassifier
---
- 사이킷런 확률적 경사 하강법을 사용하는 분류 모델
- 미니배치 경사 하강법, 배치 경사 하강법 지원 안함
- 신경망에서는 미니배치 경사 하강법 지원함
- loss 매개변수로 손실 함수 지정
- 로지스틱 회귀 `log`, 서포트 벡터 머신 `hinge`
- penalty 매개변수로 규제 종류 지정, 기본값 `l2`
- alpha 매개변수 규제 강도 지정, 기본값 `0.0001`
- max_iter 매개변수 에포크 횟수 지정, 기본값 `1000`
- tol 매개변수 반복을 멈출 조건, 기본값 `0.001`
- n_iter_no_change 매개변수에서 지정한 에포크 동안 손실이 tol 만큼 줄어들지 않으면 학습 중단, 기본값 5

## SGDSGDRegress
---
- 사이킷런 확률적 경사 하강법을 사용하는 회귀 모델
- loss 매개변수 기본값은 제곱 오차 `squared_loss` 
- 나머지 매개변수는 동일

```python
# 데이터 준비
import pandas as pd

fish = pd.read_csv('https://bit.ly/fish_csv_data')
```

```python
# 입력 데이터와 타깃 준비
fish_input = fish[['Weight','Length','Diagonal','Height','Width']].to_numpy()
fish_target = fish['Species'].to_numpy()
```

```python
# 훈련 세트와 테스트 세트 분할
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    fish_input, fish_target, random_state=42)
```

```python
# 표준화 전처리, 꼭 훈련 세트 통계 값으로 테스트 세트 변환 
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
# SGDClassifier
from sklearn.linear_model import SGDClassifier
```

```python
import sklearn
sklearn.__version__
```

```python
# 사이킷런 1.1.0 버전 이하일 경우 'log_loss'를 'log'로 바꾸어 주세요.
sc = SGDClassifier(loss='log', max_iter=10, random_state=42)
sc.fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

### 점진적 학습
---
- partial_fit() 메서드 사용
- 호출할 때마다 1 에포크씩 이어서 훈련

```python
sc.partial_fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

## 에포크와 과대/과소적합
---
- 에포크 횟수에 따라 과대/과소 적합 됨
- 훈련 세트 정확도는 증가하고 테스트 세트 정확도가 감소하는 지점에서 훈련 멈추기 => 조기 종료 early stopping


image.png

```python
import numpy as np

# 사이킷런 1.1 이상을 사용하는 경우 loss='log'를 loss='log_loss'로 변경하세요.
sc = SGDClassifier(loss='log', random_state=42)

train_score = []
test_score = []

# partial_fit() 만으로 훈련 하려면 classes 매개변수에 타깃 값 전달
classes = np.unique(train_target)
classes
```

```python
# 300 에포크 진행, 점수 계산, `_` 버리는 값 저장하는 특별한 변수
for _ in range(0, 300):
    sc.partial_fit(train_scaled, train_target, classes=classes)
    
    train_score.append(sc.score(train_scaled, train_target))
    test_score.append(sc.score(test_scaled, test_target))
```

```python
# 정확도 그래프로 확인, 파란색 훈련 세트, 주황색 테스트 세트
# 100 에포크 이후 점수 차가 벌어짐
import matplotlib.pyplot as plt

plt.plot(train_score)
plt.plot(test_score)
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.show()
```

```python
# 사이킷런 1.1 이상을 사용하는 경우 loss='log'를 loss='log_loss'로 변경하세요.
# 최적 100 에포크로 훈련
sc = SGDClassifier(loss='log', max_iter=100, tol=None, random_state=42)
sc.fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

```python
# 사이킷런 1.1 이상을 사용하는 경우 loss='log'를 loss='log_loss'로 변경하세요.
# 손실함수 변경 후 최적 100 에포크로 훈련
sc = SGDClassifier(loss='hinge', max_iter=100, tol=None, random_state=42)
sc.fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```
